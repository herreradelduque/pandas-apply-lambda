```python
import pandas as pd
import hashlib as hl
```

```python
df_movies = pd.read_csv('data/input/IMDB-Movie-Data.csv')
```

```python
df_movies.head()
```

```python
def warm_up(x):
    if x >= 0 and x <= 999:
        return 1
    elif x >= 1000 and x <= 9999:
        return 2
    elif x >= 10000 and x <= 99999:
        return 3
    elif x >= 100000 and x <= 999999:
        return 4
    elif x >= 1000000:
        return 5
    else:
        return 0
```

```python
df_movies['bin'] = df_movies['Votes'].apply(warm_up)
df_movies.head()
```

```python
df_movies['bin'] = df_movies.apply(lambda x: warm_up(x['Votes']), axis=1 )
df_movies.head()
```

```python
df_movies['revenue per minute'] = df_movies['Revenue (Millions)'] / df_movies['Runtime (Minutes)'] 
df_movies.head()
```

```python
df_movies['revenue per minute'] = df_movies.apply(lambda x: x['Revenue (Millions)'] / x['Runtime (Minutes)'], axis=1)
df_movies.head()
```

```python
df_movies['Genre'].unique()
```

```python
def ranking(x, y):
    result = x
    
    if 'comedy' in y.lower():
        result -= 1
    if 'thriller' in y.lower():
        result += 1
    
    return result
```

```python
df_movies['new ranking'] = df_movies.apply(lambda x: ranking(x['Rank'], x['Genre']), axis=1)
df_movies.head()
```

```python
def prime_number(x, y):
    n = int(sum([ord(i) for i in x]) / y)
    
    if n == 0:
        return False
    
    for i in range(2,n):
        if (n%i) == 0:
            return False
        
    return True
```

```python
df_movies['prime number'] = df_movies.apply(lambda x: prime_number(x['Title'], x['Votes']), axis=1)
df_movies.head()
```

```python
def ranking_free(rat, vot, bincol, votsum):
    result = round(rat * vot / votsum * 100)
    if bincol == 1:
        if result >= 0 and result < 5:
            return 1
        elif result >= 5 and result <= 10:
            return 2
    elif bincol == 2:
        if result >= 0 and result < 5:
            return 3
        elif result >= 5 and result <= 10:
            return 4
    if bincol == 3:
        if result >= 0 and result < 5:
            return 5
        elif result >= 5 and result <= 10:
            return 6
    if bincol == 4:
        if result >= 0 and result < 5:
            return 7
        elif result >= 5 and result <= 10:
            return 8
    if bincol == 5:
        if result >= 0 and result < 5:
            return 9
        elif result >= 5 and result < 10:
            return 10
    
    return result
```

```python
df_movies['ranking free'] = df_movies.apply(lambda x: ranking_free(x['Rating'], x['Votes'], x['bin'], df_movies['Votes'].sum()), axis=1)
df_movies.head()
```

```python
def freaky_bonus(des, rev, vot):
    des_encode = des.encode('utf-8')
    des_hash = hl.sha256(des_encode).hexdigest()
    result = sum([int(i) for i in des_hash if i.isdigit()])
        
    if result >= rev and result <= vot:
        return True
    else:
        return False
```

```python
df_movies['freaky bonus'] = df_movies.apply(lambda x: freaky_bonus(x['Description'], x['Revenue (Millions)'], x['Votes']), axis=1)
df_movies.head()
```
